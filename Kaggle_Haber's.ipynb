```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['Surv_status'].value_counts()
```

```python
plt.subplot(221)
sns.distplot(df['Age'])

plt.subplot(222)
sns.distplot(df['Op_year'])
plt.subplot(223)
sns.distplot(df['axil_nodes'])
plt.show()
```

These are the probability density functions. We can see the age is more at 50's, and the operation year is variating but it is between 55 and 70. The number of axil_nodes are zero in most cases but its widely spread.


```python
sns.distplot(df['Age'],
             hist_kws=dict(cumulative=True),
             kde_kws=dict(cumulative=True))
```

```python
sns.FacetGrid(df, hue='Surv_status',size=4).map(sns.distplot,"axil_nodes").add_legend();
plt.show()
```

```python
sns.set_style("whitegrid");
sns.pairplot(df, hue="Surv_status", height=3);
plt.show()
```

```python
sns.boxplot(x='Surv_status',y='axil_nodes', data=df)
```

```python
sns.violinplot(x='Surv_status',y='axil_nodes',data=df,size=9)
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df,hue="Surv_status",height=4).map(plt.scatter,"Op_year","Surv_status").add_legend();
plt.show()
```
